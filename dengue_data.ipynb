```python
import pandas as pd
```

```python
df_dengue = pd.read_csv("data/dengue_data.csv")
```

```python
df_dengue
```

```python
#Filtrar aÃ±os 2021, 2022 y 2023
df_dengue = df_dengue[df_dengue["ano"].isin([2021,2022,2023])]
df_dengue
```

```python
#Convertir todos los tipos de edad a aÃ±os
from utils import convertir_a_anios
df_dengue.loc[:, ["edad_anios"]] = df_dengue.apply(convertir_a_anios, axis=1)
df_dengue
```

```python
#Filtrar a menores de 13 aÃ±os
df_dengue = df_dengue[df_dengue['edad_anios'] < 13]
df_dengue
```

```python
#Clasificar grupo de edad
#Menor a 1 aÃ±o: BebÃ©
#Menor a 5 aÃ±os: PequeÃ±o
#Entre 5 y 12 aÃ±os: Mayor
from utils import clasificar_grupo_edad
df_dengue.loc[:, ["grupo_edad"]] = df_dengue["edad_anios"].apply(clasificar_grupo_edad)
df_dengue
```

```python
#Cargar nuestro archivo de coordenadas
df_coordenadas = pd.read_csv("data/distritos_peru.csv")
#Extraer columnas Latitud y Longitud
df_coordenadas[["latitud", "longitud"]] = df_coordenadas["Geo Point"].str.split(",", expand=True)
#Convertir coordenadas a float
df_coordenadas["latitud"] = df_coordenadas["latitud"].astype(float)
df_coordenadas["longitud"] = df_coordenadas["longitud"].astype(float)
#Realizaremos el merge con datos
df_dengue = df_dengue.merge(
    df_coordenadas,
    left_on=["ubigeo"],
    right_on=["UBIGEO"],
    how="left"
)
df_dengue
```

```python
#Filtrar columnas a guardar
#Filtrar columnas a guardar
df_dengue = df_dengue[["departamento", "provincia", "distrito", "ubigeo", "latitud","longitud", "enfermedad", "ano", "tipo_dx", "edad_anios", "grupo_edad", "sexo"]]
#Guardar archivo
df_dengue.to_csv("data/dengue_ninos.csv", index=False)

df_dengue
```
